```python
# Connect to MongoDB Atlas
client = pymongo.MongoClient(CONNECTION_URL)
data_base = client[DB_NAME]
```

```python
def insert_dataset(COLLECTION_NAME, df: pd.DataFrame):
    collection = data_base[COLLECTION_NAME]
    records = df.to_dict("records")
    collection.delete_many({})  # Clear collection before inserting new data
    result = collection.insert_many(records)  # <- Store result
    print(f"Inserted {len(records)} records into '{COLLECTION_NAME}' collection.")
    
    # Print the type of result
    print(f"Type of result returned by insert_many for '{COLLECTION_NAME}': {type(result)}")
```

```python
insert_dataset("diabetes_data", diabetes_df)
insert_dataset("heart_data", heart_df)
insert_dataset("kidney_data", kidney_df)
```

```python
# Function to fetch data as DataFrame
def load_dataset_as_df(COLLECTION_NAME):
    collection = data_base[COLLECTION_NAME]
    data = list(collection.find())
    
    # # Remove the '_id' field if not needed
    # for record in data:
    #     record.pop('_id', None)
    
    df = pd.DataFrame(data)
    return df
```

```python
# Load each dataset
diabetes_df = load_dataset_as_df("diabetes_data")
heart_df = load_dataset_as_df("heart_data")
kidney_df = load_dataset_as_df("kidney_data")
```
